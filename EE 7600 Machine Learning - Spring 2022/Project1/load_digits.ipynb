# Train and Test

## K-nearest-neighbors:

```python
## For normalizing the dataset, we use the three lines below. Otherwise only the next line to them will be used.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=123)

from sklearn.preprocessing import scale, normalize, minmax_scale
scaled_train = minmax_scale(X_train, feature_range=(0, 1))
X_train, y_train = scaled_train, y_train

scaled_test = minmax_scale(X_test, feature_range=(0, 1))
X_test, y_test = scaled_test, y_test

# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123, stratify=y)

from sklearn.neighbors import KNeighborsClassifier

neighbors = np.arange(1,100)
train_error = np.empty(len(neighbors))
test_error = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    knn_model = KNeighborsClassifier(n_neighbors=k, metric='minkowski', weights='uniform', p=2)
    knn_model.fit(X_train, y_train)
    train_error[i] = 1 - knn_model.score(X_train, y_train)
    test_error[i] = 1 - knn_model.score(X_test, y_test)

min_error = np.amin(test_error)    
best_k = np.where(test_error == np.amin(test_error))
print('minimum error is ', min_error, 'when k is equal to', best_k, 'and the training error is', train_error[best_k])    

f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)      
plt.plot(neighbors, test_error, label = 'Testing Error')
plt.plot(neighbors, train_error, label = 'Training Error')
plt.legend()
plt.title('KNN classifier, Load Digits data set is normalized, p=2')
plt.xlabel('number of Neighbors')
plt.ylabel('Error')
plt.show()
```

### Decision tree classifier:

```python
## For normalizing the dataset, we use the three lines below. Otherwise only the next line will be used.
# from sklearn.preprocessing import scale, normalize, minmax_scale
# scaled_data = minmax_scale(data.data, feature_range=(0, 1))
# X, y = scaled_data, data.target

# # X, y = data.data, data.target

# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123, stratify=y)

from sklearn.tree import DecisionTreeClassifier

depth = np.arange(1,30)
train_error = np.empty(len(depth))
test_error = np.empty(len(depth))

for i,k in enumerate(depth):
    dtc_model = DecisionTreeClassifier(criterion = 'entropy', max_depth=k)
    dtc_model.fit(X_train, y_train)
    train_error[i] = 1 - dtc_model.score(X_train, y_train)
    test_error[i] = 1 - dtc_model.score(X_test, y_test)

min_error = np.amin(test_error)    
best_k = np.where(test_error == np.amin(test_error))
print('minimum error is ', min_error, 'when depth is equal to', best_k, 'and the training error is', train_error[best_k])

f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)      
plt.plot(depth, test_error, label = 'Testing Error')
plt.plot(depth, train_error, label = 'Training Error')
plt.legend()
plt.title('Decision tree classifier, Load Digits data set is normalized, criterion=entropy')
plt.xlabel('number of depth')
plt.ylabel('Error')
plt.show()
```
